### The business intelligence terms and technical point

***How do we measure business performance?***

It should start by collecting observations.
    
    Can observe variables such as sales volume, new customers, each of this is considered a single observation.

No mathematical manipulations can be applied to these observations, it must do is quantify that information.

    The quantification is the process of representing observations as numbers.
    
The measure is the accumulation of observations to show more information.
___
### The frequently used terms

***Metric*** refers to a value that derives from the measures you obtain and aims at gauging business performance or progress to compare.

So if the measure is related to something like simple descriptive statistics of past performance, a metric has a business meaning attached, and the metrics are very useful of comparisons.

***KPIs***  what you need to do is choose the metrics that are tightly aligned with your business objectives.

KPI mean key performance indicators, key are related to your main business goals, performance can show how successfully you have performed within a specified timeframe, indicators means values or metrics that indicate something related to your business performance.

***metric & KPI different***
 
The metric could indicate the traffic of a page from your website that was visited by any type of user.

The KPI instead would show the volume of the same traffic however generated only from users who have clicked on the link provided in your ad compaign.
___
### Another rules

Every quantitative meaning you can extract must be visualized.

General managers will want dashboards and reports with graphs, diagrams, maps and other easily digestible visuals supported by the most relevant numbers.
___

So with the observations, quantifications, measures, metric and the KPI into easily understood and comfortable visualizations is an important part of the business intelligence analysts job.

But often dashboards are created about KPI only, KPI dashboard looks similar to the ones containing metrics, but it can allow executives and managers to stay focused on the main business objectives.
